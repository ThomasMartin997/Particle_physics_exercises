```python
'''df = pd.read_csv("your_file.csv", dtype=str)

df = pd.read_csv("your_file.csv", dtype=str)

df = df.apply(pd.to_numeric, errors='coerce')
data = pd.read_csv("data.csv", na_values=[""])'''

'''
file_handle = pd.read_csv("file.csv", dtype=str)
file-handle = file_handle.apply(pd.to_numeric, errors='co
'''

def factori(num: int) -> int:
    fact_res = 1
    for i in range (1, num+1):
        fact_res = 1*i*fact_res
    return fact_res

def n_perm_g(num: int) -> list[int]:
    '''Takes in a positive integer and generates all permutations of of integers from 1, upto and including that integer. 
    The permutations are of the given length of the input integer with non repeating.
    The function then returns a list of all tuples containing the integer permutations where the permutation inside the tuple is a string of integers.'''
    #The amount of ways to arrange num distinct objects is naturally num * (num -1)  * (num - 2 ) *.... (num - (num-1) ) 
    # by the multiplication principle of independent objects.
    if num == 1:
        return [1]
    elif num < 1:
        return []
    elif num  == 2:
        return [[1,2],[2,1]]
    
    else:
        perm = list(range(1, num + 1))
        perm_list = [perm[:]]
        perm_count = factori(num)
        count = 0
        while count != perm_count:
            index = num - 2 #second last index to always start at
            while index >= 0 and perm[index] >= perm[index + 1]: #Start at comparing the final two elements. Decrease if not found to get largest such index where p_index < p_index+1
                index -= 1
            if index == -1: #Ie. n(n-1)...(n-(n-1)(n-(n-0)) final element
                break
            j = num - 1 #Very last index. So we have the highest index with a value higher to the right. Then we find the smallest of those right values greater than it.
            while perm[j] <= perm[index]: 
                j -= 1 #get to the right most p_j that is greater than p_i. 
            min_j = j
            for rights in range(j-1, index, -1):  # Start from j-1 and search backward to i
                if perm[rights] > perm[index] and perm[rights] < perm[min_j]:
                    min_j = rights
            
            perm[index], perm[min_j] = perm[min_j], perm[index]
            perm[index + 1:] = perm[index + 1:][::-1]
            perm_list.append(perm[:])
            count += 1
    
        return perm_list
        
strn = n_perm_g(3)
strn
```
